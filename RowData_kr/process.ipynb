```python
import pandas as pd

# CSV 파일 읽기
file_path = '기온_변환.csv'
data = pd.read_csv(file_path)

# 결측값 처리: 세종특별자치시의 기온이 결측치인 경우 대전광역시 값으로 채움
for year in data['연도'].unique():
    # 세종특별자치시의 결측값 확인
    condition = (data['행정구역별(1)'] == '세종특별자치시') & (data['연도'] == year) & (data['기온'] == '-')
    if condition.any():
        # 대전광역시의 해당 연도 기온 값 가져오기
        daejeon_value = data.loc[(data['행정구역별(1)'] == '대전광역시') & (data['연도'] == year), '기온'].values[0]
        # 세종특별자치시의 결측값을 대전광역시 값으로 채움
        data.loc[condition, '기온'] = daejeon_value

# 기온 데이터를 float로 변환
data['기온'] = data['기온'].astype(float)

# 결과 저장
output_file_path = '기온_변환_결측치처리.csv'
data.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"결측값이 처리된 데이터가 {output_file_path}에 저장되었습니다.")
```

```python
import pandas as pd

# CSV 파일 읽기
file_path = '미곡생산량_백미.csv'
df = pd.read_csv(file_path, header=1)  # 2행(0-based index 1)을 헤더로 설정

# 2행 값이 '합계:생산량 (톤)'과 '논벼:10a당 생산량 (kg)'인 열만 선택
columns_to_keep = ['합계:생산량 (톤)', '논벼:10a당 생산량 (kg)']
df = df[columns_to_keep]

# 수정된 데이터 저장
output_file_path = '미곡생산량_수정.csv'
df.to_csv(output_file_path, index=False, encoding='utf-8-sig')

print(f"필요한 열만 남긴 데이터가 {output_file_path}에 저장되었습니다.")
```
